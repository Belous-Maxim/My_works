````python
def get_page(page):
    page_link = 'https://www.cian.ru/cat.php?deal_type=sale&engine_version=2&offer_type=flat&p={}&region=1&room1=1'.format(page)
    full_page = requests.get(page_link, headers={'User-Agent': UserAgent().chrome})                                                                                                                                                                                                                                                                                                                                                                                                                                                                       
    soup = BS(full_page.content, 'html.parser')
    info = []
    infa_flats = soup.findAll('article', {'data-name': 'CardComponent', 'class': '_93444fe79c--container--2pFUD _93444fe79c--cont--1Ddh2'})
    for flat in infa_flats:
        info.append(flat.a.get('href'))
    print(full_page)
    return info
````
